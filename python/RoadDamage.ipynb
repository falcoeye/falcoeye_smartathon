```python
# visualizing the results
import pandas as pd
import plotly.express as px

column = "normalized_length"
range_color = [0,5]
df = pd.read_csv("data/riyadh/riyadh_with_gps.csv")
fig = px.density_mapbox(df, lat='latitude', lon='longitude', z=column,
                        mapbox_style="open-street-map", width=920, height=720,range_color=range_color)
fig.update_layout(
    autosize=True, hovermode='closest',
    mapbox = {'zoom': 15},
    showlegend = False)
fig.show()
```

```python
# loading the data
import pandas as pd
df = pd.read_csv("data/riyadh_dammam.csv")
```

```python
column = "normalized_length"
width,height = 920,720
layout = dict(autosize=True, hovermode='closest',
    mapbox = {'zoom': 7},
    showlegend = False)
```

```python
# visualizing the Longitudinal Cracks size Density
range_color = [0,100]
df_1 = df[df["category"]==1].reset_index(drop=False)
fig = px.density_mapbox(df_1, lat='latitude', lon='longitude', z=column,
                        mapbox_style="open-street-map", width=width, height=height,range_color=range_color)
fig.update_layout(**layout)
fig.show()
```

```python
# visualizing the Transverse Cracks size Density
range_color = [0,10]
df_2 = df[df["category"]==2].reset_index(drop=False)
fig = px.density_mapbox(df_2, lat='latitude', lon='longitude', z=column,
                        mapbox_style="open-street-map", width=width, height=height,range_color=range_color)
fig.update_layout(**layout)
fig.show()
```

```python
# visualizing the Alligator Cracks size Density
range_color = [0,50]
df_3 = df[df["category"]==3].reset_index(drop=False)
fig = px.density_mapbox(df_3, lat='latitude', lon='longitude', z=column,
                        mapbox_style="open-street-map", width=width, height=height,range_color=range_color)
fig.update_layout(**layout)
fig.show()
```

```python
# visualizing Patholes size Density
range_color = [0,2]
df_4 = df[df["category"]==4].reset_index(drop=False)
fig = px.density_mapbox(df_4, lat='latitude', lon='longitude', z=column,
                        mapbox_style="open-street-map", width=width, height=height,range_color=range_color)
fig.update_layout(**layout)
fig.show()
```
